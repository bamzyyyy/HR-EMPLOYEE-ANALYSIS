## Business question

* Which departments experience the highest employee turnover?

```python
#Create a boolean mask for employees who have exited (not Active)
HR_df['Is_Exited'] = HR_df['Exit_Status'] != 'Active'

#  Group by department to calculate total employees and exited employees
turnover_summary = HR_df.groupby('Department').agg(
    Total_Employees=('Exit_Status', 'count'),
    Turnover_Count=('Is_Exited', 'sum')
)

# Calculate the Turnover Rate (%)
turnover_summary['Turnover_Rate_%'] = (turnover_summary['Turnover_Count'] / turnover_summary['Total_Employees']) * 100

# Sort from highest turnover rate to lowest
turnover_summary = turnover_summary.sort_values(by='Turnover_Rate_%', ascending=False)

turnover_summary
```

```python


# Set figure size and grid style
plt.figure(figsize=(9, 5.5))
sns.set_theme(style="whitegrid")

# Create the bar plot using your turnover_summary DataFrame
# Note: turnover_summary.index holds the 'Department' names if grouped as above
ax = sns.barplot(
    x=turnover_summary.index, 
    y=turnover_summary['Turnover_Rate_%'], 
    palette='Reds_r'
)

# Annotate the exact percentages right above each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(
        f"{height:.2f}%", 
        (p.get_x() + p.get_width() / 2., height), 
        ha='center', va='center', 
        xytext=(0, 8), 
        textcoords='offset points',
        fontweight='bold',
        color='#333333'
    )

# Chart labels and spacing
plt.title('Employee Turnover Rate by Department', fontsize=14, pad=20, fontweight='bold')
plt.xlabel('Department', fontsize=12, labelpad=10)
plt.ylabel('Turnover Rate (%)', fontsize=12, labelpad=10)
plt.ylim(0, 11)  # Gives clean breathing room above the highest bar

plt.tight_layout()
plt.show()
```

- Are payroll expenses aligned with employee performance?

```python
# Calculate average salary for each performance tier
payroll_alignment = HR_df.groupby('Performance_Rating').agg(
    Employee_Count=('Employee_ID', 'count'),
    Average_Gross_Salary=('Gross_Salary_NGN', 'mean'),
    Average_Performance_Score=('Performance_Score', 'mean')
).sort_values(by='Average_Performance_Score', ascending=False)

# Format for clean viewing
payroll_alignment['Average_Gross_Salary'] = payroll_alignment['Average_Gross_Salary'].map('{:,.2f} NGN'.format)
payroll_alignment
```

- Which teams may be at risk of employee burnout?

```python
# Group by department to analyze burnout indicators
burnout_analysis = HR_df.groupby('Department').agg(
    Total_Employees=('Employee_ID', 'count'),
    Avg_Leave_Taken=('Days_Leave_Taken', 'mean'),
    Avg_Leave_Remaining=('Days_Leave_Remaining', 'mean'),
    Avg_Training_Hours=('Training_Hours_2024', 'mean'),
    Avg_Performance_Score=('Performance_Score', 'mean')
).sort_values(by='Avg_Leave_Remaining', ascending=True) # Teams with the least leave remaining rise to the top

burnout_analysis
```

```python
# Set up the plotting environment
sns.set_theme(style="whitegrid")
fig, ax1 = plt.subplots(figsize=(10, 6))

# 1. Primary Y-Axis: Bar chart for Leave Remaining (Shorter bars = Higher burnout risk)
# We sort ascending so the highest-risk teams appear first on the left
burnout_analysis_sorted = burnout_analysis.sort_values(by='Avg_Leave_Remaining', ascending=True)

color_bars = '#e74c3c'  # Soft red to signal risk
ax1.set_xlabel('Department', fontsize=12, labelpad=10)
ax1.set_ylabel('Average Leave Days Remaining (Lower = Higher Risk)', color=color_bars, fontsize=12)
sns.barplot(x=burnout_analysis_sorted.index, y='Avg_Leave_Remaining', data=burnout_analysis_sorted, ax=ax1, color=color_bars, alpha=0.65)
ax1.tick_params(axis='y', labelcolor=color_bars)

# 2. Secondary Y-Axis: Line chart for Training Hours
ax2 = ax1.twinx()  
color_line = '#2c3e50'  # Dark slate for distinction
ax2.set_ylabel('Average Training Hours 2024', color=color_line, fontsize=12)
sns.lineplot(x=burnout_analysis_sorted.index, y='Avg_Training_Hours', data=burnout_analysis_sorted, ax=ax2, color=color_line, marker='o', linewidth=2.5, sort=False)
ax2.tick_params(axis='y', labelcolor=color_line)

# Title and Layout adjustments
plt.title('Burnout Risk Indicators by Department', fontsize=14, pad=20, fontweight='bold')
fig.tight_layout()

# Display the chart
plt.show()
```

Based on the visual data, the organization faces two entirely distinct profiles of burnout risk across departments: Operational Fatigue in Marketing and Operations, where employees have the lowest rest buffers remaining (~7.6 and ~8.0 days) despite having the lowest training hours, indicating that their exhaustion is likely driven by heavy daily operational workloads, and Cognitive Overload in Finance and IT, which carry the heaviest training workloads (nearly 28 and 23 average hours) while holding the healthiest leave balances (over 10–11 days remaining), indicating a high risk of future burnout as they absorb massive professional development demands while delaying actual rest. To address these risks, an analyst should recommend that leadership mandate rest for Finance and IT to ensure they use their deferred leave before facing performance drops, while simultaneously auditing capacity and staffing structures in Marketing and Operations to identify why core operational demands are draining their vacation balances.

- Does employee training improve performance?

```python
# Segment training hours into 3 even tiers (Low, Medium, High)
HR_df['Training_Tier'] = pd.qcut(HR_df['Training_Hours_2024'], q=3, labels=['Low Training', 'Medium Training', 'High Training'])

# Calculate the average performance score for each tier
training_effectiveness = HR_df.groupby('Training_Tier').agg(
    Employee_Count=('Employee_ID', 'count'),
    Avg_Training_Hours=('Training_Hours_2024', 'mean'),
    Avg_Performance_Score=('Performance_Score', 'mean')
)

training_effectiveness
```

```python
plt.figure(figsize=(8, 5))
sns.set_theme(style="whitegrid")

# Create a box plot to show distribution of performance scores across tiers
sns.boxplot(
    x='Training_Tier', 
    y='Performance_Score', 
    data=HR_df, 
    palette='Greens',
    order=['Low Training', 'Medium Training', 'High Training']
)

plt.title('Distribution of Performance Scores by Training Tier', fontsize=14, pad=20, fontweight='bold')
plt.xlabel('Training Tier', fontsize=12, labelpad=10)
plt.ylabel('Performance Score (0-100)', fontsize=12, labelpad=10)
plt.tight_layout()
plt.show()
```

Based on my observation, employee training directly and significantly improves performance, as demonstrated by the clear upward shift in performance scores across the training tiers. The median line rises steadily from around 58 in the "Low Training" tier to nearly 78 for "Medium Training," and peaks at approximately 90 for the "High Training" group. Furthermore, increasing the volume of training remarkably stabilizes performance metrics; while the "Low Training" group exhibits massive data variance with scores stretching drastically from 0 to over 70, the "High Training" group tightens significantly into a narrow, reliable high-performance band where every single employee scored above an 80. This progressive lift and shrinking variance provide definitive statistical proof that the organization's training initiatives are successfully and predictably upskilling the workforce.

- Are there signs of gender pay gaps or representation imbalance?

```python
# 1. Overall Headcount and Pay Gap Analysis
gender_pay_gap = HR_df.groupby('Gender').agg(
    Employee_Count=('Employee_ID', 'count'),
    Percentage_Share=('Employee_ID', lambda x: (len(x) / len(HR_df)) * 100),
    Avg_Gross_Salary=('Gross_Salary_NGN', 'mean'),
    Avg_Performance_Score=('Performance_Score', 'mean')
)

# Format currency and percentages for clear reporting
gender_pay_gap['Percentage_Share'] = gender_pay_gap['Percentage_Share'].map('{:.1f}%'.format)
gender_pay_gap['Avg_Gross_Salary'] = gender_pay_gap['Avg_Gross_Salary'].map('{:,.2f} NGN'.format)

print("--- Overall Gender Representation & Pay Gap ---")
print(gender_pay_gap)


# 2. Departmental Representation Balance (Checking for occupational segregation)
dept_gender_split = pd.crosstab(HR_df['Department'], HR_df['Gender'], margins=True, margins_name="Total")
dept_gender_pct = pd.crosstab(HR_df['Department'], HR_df['Gender'], normalize='index') * 100

print("\n--- Departmental Gender Split (Percentages) ---")
print(dept_gender_pct.round(1).astype(str) + '%')
```

```python


# Create a figure with two distinct subplots side-by-side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
sns.set_theme(style="whitegrid")

# 1. VISUAL A: Overall Headcount Share (Pie Chart)
# Replace these values with your actual counts from gender_pay_gap
gender_counts = HR_df['Gender'].value_counts() 
colors = ['#2b5c8f', '#d95f02'] # Clean, professional contrast colors

ax1.pie(
    gender_counts, 
    labels=gender_counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=colors,
    textprops={'fontsize': 12, 'fontweight': 'bold'}
)
ax1.set_title('Overall Workforce Gender Share', fontsize=14, pad=20, fontweight='bold')

# 2. VISUAL B: Equal Pay vs Performance Analysis (Grouped Bar Chart)
# This chart allows stakeholders to see if a salary gap matches a performance gap
# First, reshape your summary data slightly to make it compatible with seaborn
summary_df = HR_df.groupby('Gender').agg(
    Avg_Salary_NGN=('Gross_Salary_NGN', 'mean'),
    Avg_Performance=('Performance_Score', 'mean')
).reset_index()

# Create a dual-axis chart for the second subplot
color_sal = '#2b5c8f'
sns.barplot(x='Gender', y='Avg_Salary_NGN', data=summary_df, ax=ax2, color=color_sal, alpha=0.85)
ax2.set_ylabel('Average Gross Salary (NGN)', color=color_sal, fontsize=12, fontweight='bold')
ax2.tick_params(axis='y', labelcolor=color_sal)
ax2.set_xlabel('Gender', fontsize=12, labelpad=10)

# Instantitate secondary axis for performance overlay
ax3 = ax2.twinx()
color_perf = '#d95f02'
sns.lineplot(x='Gender', y='Avg_Performance', data=summary_df, ax=ax3, color=color_perf, marker='o', markersize=10, linewidth=3)
ax3.set_ylabel('Average Performance Score (0-100)', color=color_perf, fontsize=12, fontweight='bold')
ax3.tick_params(axis='y', labelcolor=color_perf)
ax3.set_ylim(0, 100)

ax2.set_title('Average Salary vs. Performance by Gender', fontsize=14, pad=20, fontweight='bold')

plt.tight_layout()
plt.show()
```

Based on observation, the organization maintains a balanced macro-level representation with females making up 52.3% of the workforce and males accounting for 47.7%, yet a clear gender pay gap exists as men earn a higher average gross salary (over 400,000 NGN) than women (approximately 385,000 NGN). This compensation disparity is analytically unjustified because the performance line reveals a complete misalignment: female employees actually maintain a higher average performance score (~77) than male employees (~73). Because higher output is resulting in lower average pay for women, this indicates that the pay gap is driven by structural inequities—such as a glass ceiling effect where men dominate senior job grades or higher initial salary negotiations—requiring an immediate internal equity audit across departments and job tiers to correct the imbalance.
